***Pie Chart***

```python
def pie_charts(df, column):# Define a function to create a pie chart
      # Get unique labels in the specified column   
    labels = df[column].unique()
    sizes = []
     # Count the occurrences of each unique label
    for i in labels:
        sizes.append(df[column].value_counts()[i])
         # Create and display the pie chart
    plt.pie(sizes, labels = labels, autopct='%1.1f%%')
```

```python
# Define a function to create multiple pie charts for categorical columns in a DataFrame
def multiple_pie_charts(df):    
    plt.figure(figsize=(20,35))# Create a large figure 
    # Select columns of data type 'object' or 'int64' (assuming they are categorical)
    features = df.select_dtypes(include=['object', 'int64']).columns
    # Remove specific columns ('customerID' and 'tenure') from the list
    features = features.difference(['customerID', 'tenure'])
    # Initialize a subplot counter
    j = 1
    for i in features:
       plt.subplot(6,3,j)# Create a subplot
       plt.title("{}".format(i))
       pie_charts(merged_df,i)# Call the pie_charts function to create a pie chart
       j += 1


 # Call the multiple_pie_charts function on the 'merged_df' DataFrame      
multiple_pie_charts(merged_df)
```
