```python
def pieCharts(merged_df):
    cols = [col for col in merged_df.columns if col not in ['SeniorCitizen', 'tenure', 'customerID', 'MonthlyCharges', 'TotalCharges']]
    fig, axes = plt.subplots(4, 4, figsize=(15, 10))

    for col, ax in zip(cols, axes.ravel()):
        category_counts = merged_df[col].value_counts()
        ax.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%')
        ax.set_title(col,fontsize=15)

    plt.show()
```

```python
def plot_categorical_pie_charts(dataframe):
    # Get a list of columns with categorical data types
    categorical_columns = dataframe.select_dtypes(include=['object', 'int','float']).columns
    
    for column in categorical_columns:
        # Count the occurrences of each unique value in the column
        value_counts = dataframe[column].value_counts()
        
        # Create a pie chart
        plt.figure(figsize=(6, 6))
        plt.pie(value_counts, labels=value_counts.index, autopct='%1.1f%%', startangle=140)
        plt.title(f'Distribution of {column}')
        plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.
        plt.show()

# Assuming merged_df is your DataFrame
plot_categorical_pie_charts(merged_df)
```
